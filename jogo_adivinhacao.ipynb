```python
import random # Biblioteca que permite o código escolher um número aleatório
print("=== JOGO DE ADIVINHAÇÃO ===")
print("Tente adivinhar o número de 1 a 10")
numero = random.randint(1,10) # Escolha do numero aleatório inteiro pelo computador
tentativas_restantes = 3 #numero de tentativas inicias

while tentativas_restantes != 0: # Enquanto as tentativas forem diferentes de zero ele executa este loop
  print(f"Você tem {tentativas_restantes} tentativas")

  #Verificação para que a escolha do usuario nao seja uma letra ou espaço em branco
  while True:
    try:
        numero_do_usuario = int(input("Escolha um número: ")) #Pede para que o usuario escolha um numero inteiro
        break  # Sai do loop se a conversão der certo
    except ValueError:
        print("Entrada inválida. Tente novamente.") # Printa entrada inválida pelo usuario e mostra pra ele o input novamente

  if numero == numero_do_usuario: # Se os numeros forem iguais ele parabeniza e para o codigo
      print("parabéns, você ganhou!")
      break
  elif numero > numero_do_usuario: # Se o numero for menor do que o do computador ele avisa o usuario
      print("Seu numero é menor do que o escolhido")
  elif numero < numero_do_usuario: # Se o numero for menor do que o do computador ele avisa o usuario
    print("Seu numero é maior do que o escolhido")

  tentativas_restantes -= 1; # Diminui a quantidade das tentativas

  if tentativas_restantes != 0: #Verifica se as tentativas já não acabaram
    print("Numero Incorreto, Tente novamente \n")
  else:
    print("Suas tentativas esgotaram, tente mais uma vez!")
```
